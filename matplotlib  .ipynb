# Line Plot 
With matplotlib, you can create a bunch of different plots in Python. The most basic plot is the line plot. 
The world bank has estimates of the world population for the years 1950 up to 2010. The years are loaded in your workspace as a list called year and the corresponding populations as a list called pop


```python
#importing matplotlib
import matplotlib.pyplot as plt
year=[1950, 1970, 1990, 2010]
pop=[2.51, 3.2, 5.16, 6.97]
plt.plot(year,pop)
plt.show()
```

```python
#to do scatter plot
plt.scatter(year,pop)
plt.show()
```

# Histogram
 The list containing data on the life expectancy for different countries in 2007, is available.To see how life expectancy in different countries is distributed, let's create a histogram

```python
#to display histogram
values=[63.828, 76.423, 72.301, 42.731, 75.32, 81.235, 79.829,]
plt.hist(values)
plt.show()
```

```python
#to display histogram with a bin vale
plt.hist(values,bins=3)
plt.show()
```

```python
#comparing two histograms with same bin value
plt.hist(values,bins=15)
plt.show()
new_values=[28.8, 55.23, 43.08, 30.02, 62.48, 69.12, 66.8, 50.94]
plt.hist(new_values,bins=15)
plt.show()
```

# Customization
It's time to customize your own plot. This is the fun part, you will see your plot come to life!

```python
#importing matplotlib
import matplotlib.pyplot as plt
year=[1950, 1970, 1990, 2010]
pop=[2.51, 3.2, 5.16, 6.97]

#Add more data
year=[1800, 1850, 1900]+year
pop=[1.0, 1.262, 1.650]+pop

plt.plot(year,pop)

#labelling x and y axis
plt.xlabel('Year')
plt.ylabel('Populations')

#adding title
plt.title('World Population Projectins')

#Changing the coordinates of y axis
plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B'])

#to display the lineplot
plt.show()
```
